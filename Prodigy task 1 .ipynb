```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
tm=pd.read_csv(r"C:\Users\D\OneDrive\Desktop\abc\population.csv")
```

```python
print(tm.head())
```

```python
grouped_data = tm.groupby("Country Name")["2021"].sum()
```

```python
top_categories = grouped_data.nlargest(5)
```

```python
# print the grouped data
```

```python
print("Grouped Data:")
print(top_categories)
```

```python
plt.figure(figsize=(16,10))
plt.bar(top_categories.index,top_categories.values)
plt.xlabel("Country name")
plt.title("Highest population top 5")
plt.show()
```

```python
# Select the top 5 countries based on the population in 2021
top_countries_2021 = tm.nlargest(5,'2021')
```

```python
#Melt the DataFrame to reshape it for visualization
melted_tm = top_countries_2021.melt(id_vars=['Country Name'], value_vars=['2021', '2022'], var_name='Year', value_name='Population')
```

```python
# Create a vertical bar graph 
plt.figure(figsize=(10,6))
plt.bar(melted_tm['Country Name'],melted_tm['Population'],color=melted_tm['Year'].apply(lambda x: 'red' if x == '2021' else 'green'))
plt.xlabel('Country Name')
plt.ylabel('Population')
plt.title('Comparison of Population for Top 5 Countries in 2021 and 2022')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns 
```

```python
tm=pd.read_csv(r"C:\Users\D\OneDrive\Desktop\abc\population.csv")
```

```python
selected_year ='2022'
plt.figure(figsize=(8,6))
sns.histplot(data=tm, x=selected_year, bins=10, kde=True)
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title(f'Population Distribution in {selected_year}')
plt.tight_layout()
plt.show()
```
